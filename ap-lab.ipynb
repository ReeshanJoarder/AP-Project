This dataset appears to be a comprehensive collection of technical indicators, price data, and volume metrics commonly used in stock market analysis. Below is an explanation of the columns and their potential use in stock market analysis:

---

### **1. Basic Price and Volume Data**
- **date**: The date of the trading data.
- **open**: The opening price of the stock for the day.
- **high**: The highest price of the stock during the day.
- **low**: The lowest price of the stock during the day.
- **close**: The closing price of the stock for the day.
- **adjclose**: The adjusted closing price, accounting for dividends, stock splits, and other corporate actions.
- **volume**: The total number of shares traded during the day.
- **ticker**: The stock ticker symbol (e.g., AAPL for Apple).

---

### **2. Relative Strength Index (RSI)**
- **RSIadjclose15**: RSI calculated on the adjusted closing price over a 15-day period.
- **RSIvolume15**: RSI calculated on the trading volume over a 15-day period.
- **RSIadjclose25**: RSI calculated on the adjusted closing price over a 25-day period.
- **RSIvolume25**: RSI calculated on the trading volume over a 25-day period.
- **RSIadjclose50**: RSI calculated on the adjusted closing price over a 50-day period.
- **RSIvolume50**: RSI calculated on the trading volume over a 50-day period.

RSI is a momentum oscillator that measures the speed and change of price movements, typically used to identify overbought or oversold conditions.

---

### **3. Moving Average Convergence Divergence (MACD)**
- **MACDadjclose15**: MACD calculated on the adjusted closing price over a 15-day period.
- **MACDvolume15**: MACD calculated on the trading volume over a 15-day period.
- **MACDsig-adjclose-15**: The signal line for MACD (9-day EMA of MACD) for the adjusted closing price over a 15-day period.
- **MACDdif-adjclose-15-0**: The difference between MACD and its signal line for the adjusted closing price over a 15-day period.
- **MACDhistadjclose15**: The MACD histogram (difference between MACD and its signal line) for the adjusted closing price over a 15-day period.

Similar columns exist for **25-day** and **50-day** periods.

MACD is a trend-following momentum indicator that shows the relationship between two moving averages of a security’s price.

---

### **4. Lagged Price and Volume Data**
- **laglow1**: The lowest price of the stock lagged by 1 day.
- **laghigh1**: The highest price of the stock lagged by 1 day.
- **lagvolume1**: The trading volume lagged by 1 day.
- Similar columns exist for **2-day**, **5-day**, **10-day**, and **15-day** lags.

These columns help analyze past price and volume movements to identify trends or patterns.

---

### **5. Fear and Greed Index**
- **feargreed**: A sentiment indicator that measures market emotions, ranging from extreme fear to extreme greed.

---

### **6. Exponential Moving Averages (EMA)**
- **emaadjclose5**: 5-day EMA of the adjusted closing price.
- **emavolume5**: 5-day EMA of the trading volume.
- Similar columns exist for **10-day**, **15-day**, and **50-day** periods.

EMA gives more weight to recent prices, making it more responsive to new information.

---

### **7. Simple Moving Averages (SMA)**
- **smaadjclose5**: 5-day SMA of the adjusted closing price.
- **smavolume5**: 5-day SMA of the trading volume.
- Similar columns exist for **10-day**, **15-day**, **25-day**, and **50-day** periods.

SMA is the average price over a specific period, used to smooth out price data.

---

### **8. Hammer Patterns**
Columns like **hammer1y1highhighhigh**, **hammer1y1highhighlow**, etc., represent candlestick patterns (e.g., hammer patterns) over different timeframes (1-year, 2-year, etc.). These patterns are used in technical analysis to predict potential reversals in price trends.

---

### **9. Volume-Weighted Average Price (VWAP)**
- **vwapadjclosevolume**: The volume-weighted average price, which gives the average price a stock has traded at throughout the day, based on both volume and price.

---

### **10. Bollinger Bands**
- **bollingerMA5-5adjclose**: 5-day moving average for the adjusted closing price.
- **bollingerBU5-5adjclose**: Upper Bollinger Band for the adjusted closing price (5-day MA + 2 standard deviations).
- **bollingerBL5-5adjclose**: Lower Bollinger Band for the adjusted closing price (5-day MA - 2 standard deviations).
- Similar columns exist for **volume** and different periods (e.g., 10-day, 30-day).

Bollinger Bands are used to measure volatility and identify overbought or oversold conditions.

---

### **11. Average True Range (ATR)**
- **atr5**: 5-day ATR, a measure of market volatility.
- **atr10**: 10-day ATR.
- **atr15**: 15-day ATR.
- **atr20**: 20-day ATR.

---

### **12. Commodity Channel Index (CCI)**
- **cci5**: 5-day CCI, a momentum-based indicator used to identify cyclical trends.
- **cci10**: 10-day CCI.
- **cci15**: 15-day CCI.
- **cci25**: 25-day CCI.
- **cci40**: 40-day CCI.
- **cci50**: 50-day CCI.

---

### **13. Supernova Indicators**
Columns like **supernovaTipoA-2**, **supernovaTipoB-1-1**, etc., appear to be custom or proprietary indicators, possibly combining multiple technical indicators to predict price movements or trends.

---

### **14. Candlestick Patterns and Relative Strength**
- **velaE**, **velarelativaE**: Candlestick patterns and their relative strength.
- **fuerzarelativaE-lag0**: Relative strength of a candlestick pattern at lag 0.
- **fuerzarelativaE-lag0-clip**: Clipped version of the relative strength at lag 0.
- Similar columns exist for **velaF** and other lags.

---

### **15. Money Flow Indicators**
- **mfm**: Money Flow Multiplier, used in Money Flow Index (MFI).
- **mfv**: Money Flow Volume.
- **adl**: Accumulation/Distribution Line, which uses volume and price to measure buying and selling pressure.

---

### **16. Stochastic Oscillator**
- **low-5**: Lowest price over a 5-day period.
- **high-5**: Highest price over a 5-day period.
- **K-5**: %K line of the stochastic oscillator over a 5-day period.
- **D-5**: %D line (3-day SMA of %K) over a 5-day period.
- **stochastic-k-5**: Stochastic %K over a 5-day period.
- **stochastic-d-5**: Stochastic %D over a 5-day period.
- **stochastic-kd-5**: Difference between %K and %D over a 5-day period.

Similar columns exist for **10-day** and **15-day** periods.

---

### **17. Volume and Price Differentials**
- **volumenrelativo**: Relative volume, comparing current volume to historical averages.
- **diff**: Price difference between two time periods.
- **INCREMENTO**: Percentage change in price or volume.

---

### **18. Target Variable**
- **TARGET**: The target variable, likely representing the future price movement or return that the model aims to predict.

---

### **Summary**
This dataset is a rich collection of technical indicators, price data, and volume metrics used for stock market analysis. It includes:
- **Price and volume data**: Basic metrics like open, high, low, close, and volume.
- **Technical indicators**: RSI, MACD, Bollinger Bands, ATR, CCI, Stochastic Oscillator, etc.
- **Candlestick patterns**: Hammer patterns and other candlestick-based indicators.
- **Custom indicators**: Supernova and other proprietary metrics.
- **Target variable**: Likely the future price movement or return.

This dataset is likely used for predictive modeling, such as forecasting stock prices or identifying trading signals.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix
import matplotlib.pyplot as plt
```

```python
x = df2.drop(columns = ['TARGET'])
y = df2['TARGET']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
acc_arr = []
f1_arr = []
recall_arr = []
pre_arr = []
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=72)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
acc_arr.append(accuracy)
f1 = f1_score(y_test, y_pred, average='weighted')
f1_arr.append(f1)
recall = recall_score(y_test, y_pred, average='weighted')
recall_arr.append(recall)
precision = precision_score(y_test, y_pred, average='weighted')
pre_arr.append(precision)

print(f'Accuracy: {accuracy:.4f}')
print(f'F1 Score: {f1:.4f}')
print(f'Recall: {recall:.4f}')
print(f'Precision: {precision:.4f}')

# Plot confusion matrix heatmap
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=set(y_test), yticklabels=set(y_test))
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix Heatmap')
plt.show()
```
